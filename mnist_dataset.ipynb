```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
train_images[0]
```

```python
train_labels[0]
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
train_images[0]
```

```python
total_classes = 10
train_vec_labels = keras.utils.to_categorical(train_labels, total_classes)
test_vec_labels = keras.utils.to_categorical(test_labels, total_classes)
```

```python
train_vec_labels[0]
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'), 
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='sgd',
    loss='mean_squared_error',
    metrics=['accuracy'])
```

```python
model.fit(train_images,train_vec_labels,epochs=7,verbose=True)
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['accuracy'])
```

```python
model.fit(train_images,train_vec_labels,epochs=7,verbose=True)
```

```python
model.compile(
    optimizer='sgd',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(train_images,train_vec_labels,epochs=7,verbose=True)
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(train_images,train_vec_labels,epochs=7,verbose=True)
```

```python
eval_loss, eval_accuracy = model.evaluate(test_images, test_vec_labels, verbose=False)
print("Model accuracy: %.2f" % eval_accuracy)
```
